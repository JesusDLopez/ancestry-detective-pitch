### Deep Dive: What is a VCF File?

A **VCF (Variant Call Format)** file is essentially a giant table where:

- **Each row** = one position in the genome where humans differ (a SNP)
- **Each column** = one person's genotype at that position

Think of it like this:

```
                    Person1   Person2   Person3   Person4   ...
Position 16050075:    A/A       A/G       A/A       G/G
Position 16050115:    G/G       G/A       G/G       G/A
Position 16050213:    C/C       C/T       C/T       C/C
...
(1.1 million rows)
```

**Key concepts:**

| Term | Meaning | Example |
|------|---------|---------|
| **REF** | Reference allele (most common) | A |
| **ALT** | Alternate allele (the variant) | G |
| **0** | Reference allele | A |
| **1** | Alternate allele | G |
| **0\|0** | Both chromosomes have REF | A/A (homozygous reference) |
| **0\|1** | One REF, one ALT | A/G (heterozygous) |
| **1\|1** | Both chromosomes have ALT | G/G (homozygous alternate) |

Let's look at a real example from our data:

````python
# Let's really understand what a VCF contains
# Look at ONE specific variant across a few people

variant_idx = 100  # Pick the 100th variant

print("="*60)
print("EXAMINING A SINGLE VARIANT (SNP)")
print("="*60)

print(f"\n📍 LOCATION:")
print(f"   Chromosome: {callset['variants/CHROM'][variant_idx]}")
print(f"   Position:   {callset['variants/POS'][variant_idx]:,}")

print(f"\n🧬 THE VARIANT:")
print(f"   Reference allele (REF): {callset['variants/REF'][variant_idx]}")
print(f"   Alternate allele (ALT): {callset['variants/ALT'][variant_idx][0]}")
print(f"   SNP ID: {callset['variants/ID'][variant_idx]}")

print(f"\n👥 GENOTYPES FOR FIRST 10 PEOPLE:")
print("-" * 50)
for i in range(10):
    sample_id = callset['samples'][i]
    genotype = callset['calldata/GT'][variant_idx, i, :]
    
    # Translate to alleles
    ref = callset['variants/REF'][variant_idx]
    alt = callset['variants/ALT'][variant_idx][0]
    
    allele1 = ref if genotype[0] == 0 else alt
    allele2 = ref if genotype[1] == 0 else alt
    
    # Additive encoding
    additive = genotype[0] + genotype[1]
    
    print(f"   {sample_id}: [{genotype[0]}|{genotype[1]}] → {allele1}/{allele2} → additive: {additive}")
````

````python
# Create a readable table view of the VCF data
# Show first 10 variants × first 8 people

def genotype_to_str(gt, ref, alt):
    """Convert [0,1] to 'A/G' format"""
    a1 = ref if gt[0] == 0 else alt
    a2 = ref if gt[1] == 0 else alt
    return f"{a1}/{a2}"

# Build a nice table
n_variants = 10
n_samples = 8

table_data = []
for v in range(n_variants):
    row = {
        'Position': f"{callset['variants/POS'][v]:,}",
        'REF': callset['variants/REF'][v],
        'ALT': callset['variants/ALT'][v][0],
    }
    for s in range(n_samples):
        sample_id = callset['samples'][s][:7]  # Truncate for display
        gt = callset['calldata/GT'][v, s, :]
        row[sample_id] = genotype_to_str(gt, row['REF'], row['ALT'])
    table_data.append(row)

vcf_table = pd.DataFrame(table_data)
print("📊 VCF DATA: First 10 variants × 8 people")
print("=" * 80)
vcf_table
````

### The Big Picture: What We Have

After loading and converting the VCF, we have a **genotype matrix**:

```
                    HG00096  HG00097  HG00099  ...  (2,504 people)
SNP at pos 16050075:    0        0        0
SNP at pos 16050115:    0        0        0
SNP at pos 16050213:    0        0        0
...
(1,103,547 SNPs)
```

**Matrix dimensions:** `1,103,547 rows × 2,504 columns`

Each cell contains **0, 1, or 2**:
- **0** = homozygous reference (most common)
- **1** = heterozygous
- **2** = homozygous alternate

**This is our input to PCA.** We'll reduce these 1.1 million columns down to just 3 numbers per person that capture their ancestry.

### Understanding the VCF Data

**What we just loaded:**

| Metric | Value | Meaning |
|--------|-------|---------|
| **Samples** | 2,504 | One column per person in the study |
| **Variants** | 1,103,547 | Positions on Chr22 where humans differ |
| **Shape** | (1.1M, 2504, 2) | variants × samples × ploidy |

**The genotype `[0 0]` means:**
- This person has **two copies** of the reference allele (one from each parent)
- Written in VCF notation as `0|0` (homozygous reference)

**Why ploidy = 2?**
Humans are diploid — we have two copies of each chromosome (one from mom, one from dad). Each position shows both copies: `[0 0]`, `[0 1]`, or `[1 1]`.

````python
# Convert genotypes to additive encoding (0, 1, 2)
if 'callset' in dir() and callset is not None:
    gt = allel.GenotypeArray(callset['calldata/GT'])
    
    # Sum alleles: 0|0 -> 0, 0|1 -> 1, 1|1 -> 2
    genotype_matrix = gt.to_n_alt()
    
    print(f"Genotype matrix shape: {genotype_matrix.shape}")
    print("(variants x samples)")
    
    print("\nFirst 5 variants, first 10 samples:")
    print(genotype_matrix[:5, :10])
    
    print("\nValue distribution:")
    unique, counts = np.unique(genotype_matrix, return_counts=True)
    for val, count in zip(unique, counts):
        pct = count / genotype_matrix.size * 100
        print(f"  {val}: {count:,} ({pct:.1f}%)")
````

### Interpreting the Genotype Distribution

The additive encoding converts diploid genotypes to a single number:

| Genotype | Encoding | Meaning | Frequency |
|----------|----------|---------|-----------|
| `0\|0` | **0** | Homozygous reference (AA) | 94.6% |
| `0\|1` or `1\|0` | **1** | Heterozygous (AB) | 3.4% |
| `1\|1` | **2** | Homozygous alternate (BB) | 2.0% |

**Key insight:** 94.6% of all genotype calls are `0` — meaning **most of the genome is identical across all humans**. Genetic variation is rare, which is why we need millions of positions to find the ~5% that differs.

This is why ancestry prediction works: the small percentage of variants that *do* differ between people carry information about their geographic origins.

## 4. Allele Frequencies & Basic QC

````python
# Calculate allele frequencies
if 'genotype_matrix' in dir():
    # Allele frequency = mean / 2 (since diploid)
    allele_freq = genotype_matrix.mean(axis=1) / 2
    
    # Minor allele frequency (MAF)
    maf = np.minimum(allele_freq, 1 - allele_freq)
    
    print(f"Allele frequency range: {allele_freq.min():.4f} - {allele_freq.max():.4f}")
    print(f"MAF range: {maf.min():.4f} - {maf.max():.4f}")
    
    # Plot MAF distribution
    fig, axes = plt.subplots(1, 2, figsize=(14, 5))
    
    axes[0].hist(maf, bins=50, color='#10b981', edgecolor='black', alpha=0.7)
    axes[0].axvline(0.01, color='red', linestyle='--', label='MAF = 0.01 threshold')
    axes[0].set_xlabel('Minor Allele Frequency (MAF)')
    axes[0].set_ylabel('Number of SNPs')
    axes[0].set_title('MAF Distribution')
    axes[0].legend()
    
    axes[1].hist(maf[maf > 0.01], bins=50, color='#3b82f6', edgecolor='black', alpha=0.7)
    axes[1].set_xlabel('Minor Allele Frequency (MAF)')
    axes[1].set_ylabel('Number of SNPs')
    axes[1].set_title('MAF Distribution (after filtering MAF > 0.01)')
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nSNPs with MAF > 0.01: {(maf > 0.01).sum()} / {len(maf)} ({(maf > 0.01).mean()*100:.1f}%)")
````

### Why MAF Filtering Matters

**Minor Allele Frequency (MAF)** is the frequency of the less common allele at a given position.

**What the histograms show:**
- **Left plot:** Most SNPs have very low MAF (rare variants) — the huge spike near 0
- **Right plot:** After filtering (MAF > 0.01), we get a more uniform distribution

**Why filter rare variants?**

| Reason | Explanation |
|--------|-------------|
| **Sequencing errors** | Very rare variants are often technical artifacts, not real biology |
| **No population signal** | If only 1 in 10,000 people have a variant, it can't distinguish populations |
| **Statistical noise** | Rare variants add noise without adding predictive power |

**The numbers:**
- Raw: **1,103,547** SNPs
- After MAF > 0.01: **193,928** SNPs (17.6% retained)
- After MAF > 0.05: **112,614** SNPs (used for PCA)

We're keeping only the **common variants** that are shared across enough people to be informative about ancestry.

## 5. Quick PCA Preview

Let's run a quick PCA on the subset to see if populations cluster.

````python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

if 'genotype_matrix' in dir():
    # Filter to common variants (MAF > 0.05 for cleaner signal)
    common_snps = maf > 0.05
    X = genotype_matrix[common_snps, :].T  # Transpose: samples x variants
    
    print(f"Matrix for PCA: {X.shape[0]} samples x {X.shape[1]} SNPs")
    
    # Handle missing values (replace -1 with column mean)
    X = np.where(X < 0, np.nan, X)
    col_means = np.nanmean(X, axis=0)
    X = np.where(np.isnan(X), col_means, X)
    
    # Standardize
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # PCA
    pca = PCA(n_components=10)
    pcs = pca.fit_transform(X_scaled)
    
    print(f"\nVariance explained by first 10 PCs:")
    for i, var in enumerate(pca.explained_variance_ratio_[:10]):
        print(f"  PC{i+1}: {var*100:.2f}%")
````

### Understanding PCA Results

**What PCA did:**
- Input: **2,504 samples × 112,614 SNPs** (a 112K-dimensional space)
- Output: **2,504 samples × 10 principal components** (reduced to 10 dimensions)

**Variance explained:**

| Component | Variance | Cumulative | What it likely captures |
|-----------|----------|------------|------------------------|
| PC1 | 9.40% | 9.4% | African vs. non-African ancestry |
| PC2 | 3.90% | 13.3% | East Asian vs. European/South Asian |
| PC3 | 1.31% | 14.6% | Finer continental structure |
| PC4+ | <1.2% | ... | Sub-population differences |

**Why is variance so low?**

This is normal for genetic data. Unlike image data where the first few PCs might capture 80%+ of variance, genetic data has many independent signals. Even 9.4% on PC1 is substantial — it means PC1 alone captures more information about population structure than any other single axis.

**The key point:** Even though PC1-PC3 only explain ~15% of total variance, they capture almost all the *ancestry-relevant* variance. The remaining 85% is mostly random variation within populations.

````python
# Match samples to populations
if 'pcs' in dir():
    # Create dataframe with PCs and population labels
    samples_list = list(callset['samples'])
    
    pca_df = pd.DataFrame({
        'sample': samples_list,
        'PC1': pcs[:, 0],
        'PC2': pcs[:, 1],
        'PC3': pcs[:, 2]
    })
    
    # Merge with panel
    pca_df = pca_df.merge(panel[['sample', 'pop', 'super_pop']], on='sample')
    
    print(f"PCA dataframe shape: {pca_df.shape}")
    pca_df.head()
````

````python
# Plot PCA
if 'pca_df' in dir():
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))
    
    colors = {'AFR': '#10b981', 'EUR': '#3b82f6', 'EAS': '#f59e0b', 'SAS': '#a855f7', 'AMR': '#ef4444'}
    
    # PC1 vs PC2
    for superpop in ['AFR', 'EUR', 'EAS', 'SAS', 'AMR']:
        mask = pca_df['super_pop'] == superpop
        axes[0].scatter(pca_df.loc[mask, 'PC1'], pca_df.loc[mask, 'PC2'], 
                       c=colors[superpop], label=superpop, alpha=0.6, s=20)
    axes[0].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
    axes[0].set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]*100:.1f}%)')
    axes[0].set_title('PCA: PC1 vs PC2')
    axes[0].legend()
    
    # PC1 vs PC3
    for superpop in ['AFR', 'EUR', 'EAS', 'SAS', 'AMR']:
        mask = pca_df['super_pop'] == superpop
        axes[1].scatter(pca_df.loc[mask, 'PC1'], pca_df.loc[mask, 'PC3'], 
                       c=colors[superpop], label=superpop, alpha=0.6, s=20)
    axes[1].set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]*100:.1f}%)')
    axes[1].set_ylabel(f'PC3 ({pca.explained_variance_ratio_[2]*100:.1f}%)')
    axes[1].set_title('PCA: PC1 vs PC3')
    axes[1].legend()
    
    plt.suptitle('Ancestry Clusters from Chromosome 22 SNPs (10K variant subset)', fontsize=14, y=1.02)
    plt.tight_layout()
    plt.show()
````

### Interpreting the PCA Plots

**What we're seeing:**

The 2D and 3D plots show each of the 2,504 individuals as a point, colored by their continental ancestry. The axes are the first 3 principal components.

**Key observations:**

| Population | Pattern | Biological explanation |
|------------|---------|----------------------|

## Next Step: Classification & Evaluation Metrics

### What We've Done So Far

We completed the **unsupervised** part of the pipeline:
- Loaded 1.1 million SNPs from Chromosome 22
- Filtered to ~112K common variants (MAF > 0.05)
- Applied PCA to reduce to 3 dimensions

### What's Missing: The Classifier

To get **quantitative metrics**, we need to add a **supervised classification** step:

```
PCA coordinates (PC1, PC2, PC3)  →  Classifier  →  Predicted ancestry
         ↓                                              ↓
    [12.5, -4.2, 8.1]           Random Forest      "EAS" (94% confidence)
```

### Metrics We Would Calculate

| Metric | Formula | What It Tells Us |
|--------|---------|------------------|
| **Accuracy** | (TP + TN) / Total | Overall % correct predictions |
| **Precision** | TP / (TP + FP) | Of those predicted as AFR, how many are actually AFR? |
| **Recall** | TP / (TP + FN) | Of all actual AFR individuals, how many did we correctly identify? |
| **F1-Score** | 2 × (Precision × Recall) / (Precision + Recall) | Harmonic mean of precision and recall |
| **Confusion Matrix** | — | Shows which populations get misclassified as which |

### How We Would Implement It

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import classification_report, confusion_matrix

# Features: PC coordinates | Labels: super_pop
X = pca_df[['PC1', 'PC2', 'PC3']]
y = pca_df['super_pop']

# Train classifier with cross-validation
clf = RandomForestClassifier(n_estimators=100, random_state=42)
scores = cross_val_score(clf, X, y, cv=5)

# Get predictions and metrics
y_pred = cross_val_predict(clf, X, y, cv=5)
print(classification_report(y, y_pred))
```
